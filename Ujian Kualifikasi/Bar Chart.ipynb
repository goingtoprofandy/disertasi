```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file Excel
data = pd.read_excel('nama_file.xlsx')

# Menghitung jumlah masing-masing jenis item per tahun
counts = data.groupby(['Publication Year', 'Item Type']).size().unstack(fill_value=0)

# Membuat plot
counts.plot(kind='bar', stacked=False, figsize=(10, 6))
plt.xlabel('Year')
plt.ylabel('Number of Items')
plt.title('Items by Year')
plt.legend(title='Item Type')
plt.show()
```
