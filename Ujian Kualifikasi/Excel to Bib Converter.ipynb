```python
import pandas as pd
import numpy as np
import re

def remove_html_tags(text):
    # Define regular expression pattern to match HTML paragraph tags
    pattern = r'<\/?p>'

    # Remove HTML paragraph tags from the text
    cleaned_text = re.sub(pattern, '', text)

    return cleaned_text

def excel_to_bib(excel_file, bib_file):
    # Baca file Excel
    df = pd.read_excel(excel_file)

    # Membuat kamus untuk memetakan SourceType ke jenis entri .bib yang sesuai
    source_type_mapping = {
        'conferencePaper': 'inproceedings',
        'journalArticle': 'article'
        # Tambahkan pemetaan untuk jenis SourceType lainnya jika diperlukan
    }

    journal_name_mapping = {
        'conferencePaper': 'booktitle',
        'journalArticle': 'journal'
        # Tambahkan pemetaan untuk jenis SourceType lainnya jika diperlukan
    }

    # Buka file .bib untuk menulis
    with open(bib_file, 'w', encoding='utf-8') as f:
        # Iterasi setiap baris dalam dataframe
        for index, row in df.iterrows():
            # Mendapatkan jenis entri .bib yang sesuai berdasarkan SourceType
            bib_type = source_type_mapping.get(row['Item Type'], 'misc')  # 'misc' digunakan jika tidak ada pemetaan yang cocok
            journal_name_type = journal_name_mapping.get(row['Item Type'], 'misc')  # 'misc' digunakan jika tidak ada pemetaan yang cocok

            # Format data sesuai dengan format .bib
            bib_entry = f"@{bib_type}{{\n"
            bib_entry += f"    abstract = {{{row['Abstract Note']}}},\n"
            bib_entry += f"    author = {{{row['Author']}}},\n"  # Ganti 'Author' dengan kolom yang sesuai dari Excel

            if (pd.isna(row['Place']) == False):
                bib_entry += f"    address = {{{row['Place']}}},\n"  # Ganti 'Author' dengan kolom yang sesuai dari Excel

            if (pd.isna(row['DOI']) == False):
                bib_entry += f"    doi = {{{row['DOI']}}},\n"

            if (pd.isna(row['ISBN']) == False):
                bib_entry += f"    isbn = {{{row['ISBN']}}},\n"  # Insert ISBN or ISSN based on the condition

            if (pd.isna(row['ISSN']) == False):
                bib_entry += f"    issn = {{{row['ISSN']}}},\n"  # Insert ISBN or ISSN based on the condition

            bib_entry += f"    {journal_name_type} = {{{row['Publication Title']}}},\n"  # Ganti 'JournalName' dengan kolom yang sesuai dari Excel

            if (pd.isna(row['Manual Tags']) == False):
                bib_entry += f"    keywords = {{{row['Manual Tags']}}}\n"

            if (pd.isna(row['Notes']) == False):
                bib_entry += f"    note = {{{remove_html_tags(row['Notes'])}}},\n"

            if (pd.isna(row['Issue']) == False):
                bib_entry += f"    number = {{{row['Issue']}}},\n"  # Ganti 'Issue' dengan kolom yang sesuai dari Excel

            if (pd.isna(row['Pages']) == False):
                bib_entry += f"    pages = {{{row['Pages']}}},\n"

            if (pd.isna(row['Publisher']) == False):
                bib_entry += f"    publisher = {{{row['Publisher']}}},\n"  # Ganti 'Issue' dengan kolom yang sesuai dari Excel
            if (pd.isna(row['Title']) == False):
                bib_entry += f"    title = {{{row['Title']}}},\n"
            if (pd.isna(row['Url']) == False):
                bib_entry += f"    url = {{{row['Url']}}},\n"
            if (pd.isna(row['Publication Year']) == False):
                bib_entry += f"    year = {{{row['Publication Year']}}},\n"
            if (pd.isna(row['Volume']) == False):
                bib_entry += f"    volume = {{{row['Volume']}}},\n"  # Ganti 'Volume' dengan kolom yang sesuai dari Excel

            if (pd.isna(row['Series']) == False):
                bib_entry += f"    series = {{{row['Series']}}},\n"  # Ganti 'Volume' dengan kolom yang sesuai dari Excel
            bib_entry += "}\n\n"

            # Tulis entri .bib ke file
            f.write(bib_entry)

# Ubahlah 'nama_file_excel.xlsx' menjadi nama file Excel yang sesuai
# Ubahlah 'hasil_konversi.bib' menjadi nama file .bib yang ingin kamu buat
excel_to_bib('nama_file_excel.xlsx', 'hasil_konversi.bib')
```
